## Loading Custom Model

## Making Detections with Images

```python
%matplotlib inline
import matplotlib.pyplot as plt
img = r"C:\Users\Hp\Desktop\smoking and cellphone detection\data\images\8.jpg" #Enter the path of the image

results = model(img)
print(results)
plt.imshow(np.squeeze(results.render()))
```

## Making Detections with videos

```python
cap = cv2.VideoCapture(r"C:\Users\Hp\Desktop\Smoke Detection\stock-footage-woman-texting-message-with-cell-phone-while-driving-car-doing-selfie-slow-motion.webm") #Enter the path of the video file
while cap.isOpened():
    ret, frame = cap.read()
    
    # Make detections 
    results = model(frame)
    
    cv2.imshow('YOLO', np.squeeze(results.render()))
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

## Making live detection

```python
cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret, frame = cap.read()
    
    # Make detections 
    results = model(frame)
    
    cv2.imshow('YOLO', np.squeeze(results.render()))
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
